```python
#logistic regression using gradient descent
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
class Logreg:
   def __init__(self,n_features,n_samples,learning_rate=0.01,n_iters=1000):
        self.lr = learning_rate
        self.n_samples=n_samples
        self.n_features=n_features
        self.n_iters=n_iters
        self.weights = np.zeros((1,n_features))
        self.bias=0
  
   def y_predict(self,x):
      y_predict=sigmoid(np.dot(self.weights,x.T)+self.bias)
      return y_predict
  
  
   def gradientdescent(self,x,y,y_predict):
        error=y_predict-y
        #print(y_predict)
        dw=(1/self.n_samples)*np.dot(error,x)
        #print('dw:',dw,dw.shape)
        db=(1/self.n_samples)*np.sum(error)
        return dw,db
        
   def updateparam(self,dw,db):
        self.weights-=self.lr*dw
        self.bias-=self.lr*db

   def fit(self,x,y):
      cost_list=[]
      for i in range(self.n_iters):
        y_predict=self.y_predict(x)
        #print('ypredict',y_predict)
        cost=(-1/self.n_samples)*(np.sum((y*np.log(y_predict))+((1-y)*np.log(1-y_predict))))
        dw,db=self.gradientdescent(x,y,y_predict)
        self.updateparam(dw,db)
        cost_list.append(cost)
      return cost_list

 
```
